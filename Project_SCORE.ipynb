```python
import pandas as pd
import re
import os, glob
from pathlib import WindowsPath
import locale
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
#to convert strings with commas to floats
```

```python
#Test block
#Read all text in folder
folder= "/content/drive/MyDrive/ProjectSCORE"
files=glob.glob(folder+".jpg")
f=[]
t=[]
for root,_,file in os.walk(folder):
    for filename in file:
        f.append(filename)
        img1 = Image.open(folder+'/'+ filename)
        text = pytesseract.image_to_string(img1, lang = 'eng')
        t.append(text)
        print(text)
        #print('-='*20)
```

```python
#read one file
player_name=str(text.split('STATISTICS')[0])
sub1="Played:"
sub2="Won:"
sub3="Lost:"
sub4="Scored:"
s1=str(re.escape(sub1))
s2=str(re.escape(sub2))
s3=str(re.escape(sub3))
s4=str(re.escape(sub4))
sub5="   "
s5=str(re.escape(sub5))
games_played=locale.atoi(re.findall(s1+"(.*)"+s2,text)[0])
wins=locale.atoi(re.findall(s2+"(.*)"+s3,text)[0])
losses=locale.atoi(re.findall(s3+"(.*)"+s4,text)[0])
goals_scored=locale.atoi(re.findall(s4+"(.*)",text)[0])
print("Player Name:",player_name)
print("Games Played:",games_played)
print("Wins:",wins)
print("Losses:",losses)
print("Goals Scored:",goals_scored)
_data=[player_name,games_played,wins,losses,goals_scored]
```

```python
folder= "/content/drive/MyDrive/ProjectSCORE"
files=glob.glob(folder+".jpg")
f=[]
t=[]
player_data=pd.DataFrame(columns=['name','position','games_played','wins','losses','goals_scored'])
for root,_,file in os.walk(folder):
    for filename in file:
        f.append(filename)
        img2 = Image.open(folder+'/'+ filename)
        text = pytesseract.image_to_string(img2, lang = 'eng')
        t.append(text)
        #print(text)
        player_name=str(text.split('STATISTICS')[0])
        player_name=re.sub('[^A-Z]', '', player_name)
        sub1="Played:"
        sub2="Won:"
        sub3="Lost:"
        if "Conceded:" in text:
          sub4="Conceded:"
        else:
          sub4="Scored:"
        sub5="BEHAVIOUR"
        s1=str(re.escape(sub1))
        s2=str(re.escape(sub2))
        s3=str(re.escape(sub3))
        s4=str(re.escape(sub4))
        s5=str(re.escape(sub5))
        if sub5 in text:
          position=re.findall(""+"(.*)"+s5,text)[0]
        else:
          position=None
        games_played=locale.atoi(re.findall(s1+"(.*)"+s2,text)[0])
        wins=locale.atoi(re.findall(s2+"(.*)"+s3,text)[0])
        losses=locale.atoi(re.findall(s3+"(.*)"+s4,text)[0])
        if "GK" in text:
          goals_scored=locale.atoi(re.findall(s4+"(.*)",text)[0])*-1
        else:
          goals_scored=locale.atoi(re.findall(s4+"(.*)",text)[0])
        _data=[player_name,position,games_played,wins,losses,goals_scored]
        player_data.loc[len(player_data)]=_data
        print("Player Name:",player_name)
        print("Position:",position)
        print("Games Played:",games_played)
        print("Wins:",wins)
        print("Losses:",losses)
        print("Goals Scored:",goals_scored)

```
